```python
x= df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms',
'Area Population']]
y= df['Price']
```

```python
x.info()
```

```python
y.head()
```

### 1. Spliting The Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state = 10)
```

```python
xtrain.info()
```

```python
xtest.info()
```

```python
ytrain.shape
```

```python
xtrain.head()
```

### 2. Creating The Machine Learning Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

### 3. Training The Model with Training Data

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_,columns=['Coef.'],index=x.columns)
```

### 4. Evaluation

```python
model.score(xtest,ytest) *100
```

```python
print('%.1f%%'%(model.score(xtest,ytest)*100))
```

```python
ypred= model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(ytest,ypred))
```

```python
ypred[0]
```

```python
100*mean_absolute_error(ytest,ypred)/ypred[0]
```

##### Accuracy = 1 - error
##### Error = 1 - Accuravy

#### Conclusion: Lets use this model, because the accuracy score is more than 90%
